## EDA

```python
columns = ['is_paid','level','subject']
for i in columns:
    print(i , "-->",df[i].unique(), "\n")
      
```

```python
df["published_timestamp"]=pd.to_datetime(df["published_timestamp"])
df["Year"]=df["published_timestamp"].dt.year
```

```python
for i in columns:
    plt.title(i)
    sns.countplot(y =i, data = df)
    
    plt.show()
```

```python
year = df.groupby(["Year"])["course_id"].count()
year = pd.DataFrame(year)
year.rename(columns = {"course_id":"Total_Courses"},inplace = True)
year.reset_index(inplace=True)
year
```

```python
sns.barplot(x="Year",y="Total_Courses",data = year)
plt.show()
```

```python
sns.histplot(x="price",data = df,kde = True);
```

```python
sns.histplot(data=df,x="num_subscribers",bins=20);
```

```python
sns.histplot(data=df,x="content_duration",bins=50);
```

```python
sns.heatmap(df.corr(), cmap="coolwarm", annot=True);
```

```python
px.box(df,x ="price",y= "subject");
```

```python
top_paid_courses = df[df["price"]!=0][["course_title","num_subscribers","subject"]].sort_values(by = "num_subscribers",ascending = False)
top_paid_courses.head()
```

```python
top_unpaid_courses = df[df["price"]==0][["course_title","num_subscribers","subject"]].sort_values(by = "num_subscribers",ascending = False)
top_unpaid_courses.head()
```

```python
plt.title("Paid Courses")
sns.countplot(y ="subject", data = top_paid_courses);
plt.show()
plt.title("UnPaid Courses ")
sns.countplot(y ="subject", data = top_unpaid_courses);
plt.show()
```
